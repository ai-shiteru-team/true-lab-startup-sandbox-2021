# Clustering

```python
import json
import numpy as np
```

```python
data = open('C:/Users/Administrator/Desktop/KNN/merge_annotations.coco.json', 'r')
data = data.read()
data = json.loads(data)
```

```python
json_list = data['annotations']
```

```python
list_1 = []
list_2 = []
list_3 = []
list_4 = []
list_5 = []
list_6 = []
list_7 = []
list_8 = []
list_9 = []
list_10 = []
```

```python
for i in range(len(json_list)):
#     print('IMG',json_list[i].get('category_id'), json_list[i].get('area'))
    if json_list[i].get('category_id') == 1:
        list_1.append(json_list[i].get('area'))
    elif json_list[i].get('category_id') == 2:
        list_2.append(json_list[i].get('area'))
    elif json_list[i].get('category_id') == 3:
        list_3.append(json_list[i].get('area'))
    elif json_list[i].get('category_id') == 4:
        list_4.append(json_list[i].get('area'))
    elif json_list[i].get('category_id') == 5:
        list_5.append(json_list[i].get('area'))
    elif json_list[i].get('category_id') == 6:
        list_6.append(json_list[i].get('area'))
    elif json_list[i].get('category_id') == 7:
        list_7.append(json_list[i].get('area'))
    elif json_list[i].get('category_id') == 8:
        list_8.append(json_list[i].get('area'))
    elif json_list[i].get('category_id') == 9:
        list_9.append(json_list[i].get('area'))
    elif json_list[i].get('category_id') == 10:
        list_10.append(json_list[i].get('area'))
```

## Number

```python
print(np.shape(list_1))
print(np.shape(list_2))
print(np.shape(list_3))
print(np.shape(list_4))
print(np.shape(list_5))
print(np.shape(list_6))
print(np.shape(list_7))
print(np.shape(list_8))
print(np.shape(list_9))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


# x = [916,684,613,612,593,552,487,484,475,474,438,431,421,418,409,391,389,388,
#     380,374,371,369,357,356,340,338,328,317,316,315,313,303,283,257,255,254,245,
#     234,232,227,227,222,221,221,219,214,201,200,194,169,155,140]

kmeans = KMeans(n_clusters=3)
model = kmeans.fit(np.reshape(list_1, (len(list_1), 1)))
centroids = kmeans.cluster_centers_

labels = kmeans.labels_

print(centroids)
print(labels)

colors = ["g.","r.","y."]

for i in centroids: plt.plot( [0, len(list_1)-1],[i,i], "k" )
for i in range(len(list_1)):
    plt.plot(i, list_1[i], colors[labels[i]], markersize = 10)

plt.show()
```
